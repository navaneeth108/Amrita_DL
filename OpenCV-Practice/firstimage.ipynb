```
#Read and Display the image
img = cv2.imread('cameraman.jpg')
img_cvt=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_cvt)
```

```
#Filtering in OpenCV
#Smoothening Kernel
SMkernel = np.ones((3, 3), np.float32) / 9
smImg = cv2.filter2D(img, -1, SMkernel)
plt.imshow(smImg)
```

```
#Sharpening Kernel
SHkernel = np.array([[0, -1, 0],
                   [-1, 5, -1],
                   [0, -1, 0]])
shImg = cv2.filter2D(img, -1, SHkernel)
plt.imshow(shImg)
```

```
#Edge Filtering 
SVkernel = np.array([[-1, -2, -1],
                   [0, 0, 0],
                   [1, 2, 1]])
SVImg = cv2.filter2D(img, -1, SVkernel)
plt.imshow(SVImg)
```

```
#There are ready made functions
#cv2.blur, cv2.Sobel, cv2.Laplacian, cv2.Canny
CEImg = cv2.Canny(img,100,200)
plt.imshow(CEImg, cmap="gray")
```

```
sobelx = cv2.Sobel(src=img, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=5)
plt.imshow(sobelx)
```

```
#Detection on the X axis

sobely = cv2.Sobel(src=img, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=5) # Sobel Edge
plt.imshow(sobely)
```

```
sobelxy = cv2.Sobel(src=img, ddepth=cv2.CV_64F, dx=1, dy=1, ksize=5) # Combined X
plt.imshow(sobelxy)
```
